```python
import multiprocessing as mp

def test():
    q = mp.Queue()
    try:
        import nonexistent_package
    except ImportError as e:
        q.put(e)  # Can we send the exception object?
    
    exc = q.get()
    print(f"Received exception: {exc}")
    print(f"Type of exception: {type(exc)}")

    return exc

exc = test()

if isinstance(exc, ModuleNotFoundError):
    print("ModuleNotFoundError was successfully sent through the multiprocessing Queue.")
```
