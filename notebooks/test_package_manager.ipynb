```python
# Test uninstalling a package
print("Testing package uninstallation...")

namespace = get_persistent_namespace()
namespace.clear()

# Ensure package is installed
package = 'numpy'
pm.install_package(package)

print(f"Uninstalling package: {package}")
pm.uninstall_package(package)
```

```python
# Test Installing a package
print("Testing package installation...")
namespace = get_persistent_namespace()
namespace.clear()

package = 'numpy'
print(f"Installing package: {package}")

pm.install_package(package)
```

```python
from sparq.tools.python_repl.package_manager import PackageManager as pm
from sparq.tools.python_repl.namespace import get_persistent_namespace
import sys

# Test uninstalling a package
print("Testing package uninstallation...")

namespace = get_persistent_namespace()
namespace.clear()

# Ensure package is installed
package = 'numpy'
pm.install_package(package)

# Import into namespace
exec("import numpy as np", namespace)
print(f"After import - 'numpy' in namespace: {'numpy' in namespace}")
print(f"After import - 'numpy' in sys.modules: {'numpy' in sys.modules}")

print(f"\nUninstalling package: {package}")
result = pm.uninstall_package(package)

print(f"After uninstall - 'numpy' in namespace: {'numpy' in namespace}")
print(f"After uninstall - 'numpy' in sys.modules: {'numpy' in sys.modules}")

# Try to check if installed
print(f"\nis_installed check: {pm.is_installed(package)}")
```
